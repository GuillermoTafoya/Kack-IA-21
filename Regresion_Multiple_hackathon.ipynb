```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_excel('tres.xlsx')
np.any(np.isnan(dataset))
np.all(np.isfinite(dataset))
```

```python

assert isinstance(dataset, pd.DataFrame), "df needs to be a pd.DataFrame"
dataset.dropna(inplace=True)
indices_to_keep = ~dataset.isin([np.nan, np.inf, -np.inf]).any(1)
dataset[indices_to_keep].astype(np.float64)
```

```python
dataset.replace('N.A.',0)
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,0].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=7, suppress=True)
coefficients = pd.concat([pd.DataFrame(dataset.columns[1:]),pd.DataFrame(np.transpose(regressor.coef_))], axis = 1)
print(coefficients)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
#concatenado=np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)
#lista = [i for i in concatenado ]
# kk=r2_score(y_test, y_pred, multioutput='variance_weighted')
# print(kk)
# # print(lista)
```

```python
print(r2_score(y_test, y_pred, multioutput='variance_weighted'))
```
